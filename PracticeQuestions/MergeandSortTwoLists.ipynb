```python
def merge_and_sort(l1,l2):
    l1.extend(l2)
    for i in range(len(l1)):
        for j in range(i+1,len(l1)):
            if l1[i]>l1[j]:
                l1[i],l1[j]=l1[j],l1[i]
    return l1
l1=list(map(int,input("enter the elements seperated by space:").split()))  
l2=list(map(int,input("enter the elements seperated by space:").split()))  
print(merge_and_sort(l1,l2))

```
