```python
def length(word):
    count = 0
    for ch in word:
        count += 1
    return count    

def find_longest(sentence):
    words = sentence.split()
    longest = words[0]
    for word in words:
        if length(word) > length(longest):
            longest = word
    return longest

sentence = "Python is fantastic and powerful"
print("Longest word:"+(find_longest(sentence)))
```
