```python
def second_large_and_small(arr):
   
    firstL = float('-inf')
    secondL = float('-inf')

    for x in arr:
        if x > firstL:
            secondL = firstL
            firstL = x
            
        elif firstL > x > secondL:
            secondL = x

    print("Second largest:", secondL)

    
    firstMin = float('inf')
    secondMin = float('inf')

    for x in arr:
        if x < firstMin:
            secondMin = firstMin
            firstMin = x
        elif firstMin < x < secondMin:
            secondMin = x

    print("Second smallest:", secondMin)


second_large_and_small([1, 2, 3, 4, 5, 6])
```
