```python
def grade(lst):
    grade=[]
    for i in lst:
        if i>=90:
            grade.append("A")
        elif i>=80:
            grade.append("B") 
        elif i>=70:
            grade.append("C")
        elif i>=60:
            grade.append("D")
        elif i>=50:
            grade.append("E")
        else:
            grade.append("F")
    return grade        
lst = [85, 40, 77, 92, 60]
print(grade(lst))                               
```
