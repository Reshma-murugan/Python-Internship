A palindrome is a word, number, or sentence that reads the same forward and backward.

Examples:

madam

level

racecar

121

radar

🔹 Method 1: Using reverse string

```python
def is_palindrome(s):
   if s==s[ : :-1]:
    return "palindrome"
   else:
    return "not a palindrome"

print(is_palindrome("amma")) 
```

```python
def is_palindrome(s):
    start=0
    end=len(s)-1
    while start<end:
        if s[start]!=s[end]:
            return "not a palindrome"
        start+=1
        end-=1
    return "palindrome"

print(is_palindrome("appa"))
```

✅ 1. Time Complexity of Palindrome Check
Method 1: Using reverse s[::-1]
s == s[::-1]

Creating s[::-1] takes O(n) time.

Comparing strings takes O(n) time.

Total:

⭐ Time Complexity = O(n)
Method 2: Using two pointers
while i < j:
    if s[i] != s[j]:
        return False

You check each character once.

Loop runs roughly n/2 times.

⭐ Time Complexity = O(n)
✅ 2. Space Complexity
Method 1: Using reverse s[::-1]

It creates a new reversed string, which uses extra memory.

Reverse string → O(n) space

⭐ Space Complexity = O(n)
Method 2: Two-pointer method

You do not create any new string.

Just two integer variables i, j

⭐ Space Complexity = O(1)

(constant space)

📌 Summary (Important for Interviews)
Method	Time	Space
Reverse string s[::-1]	O(n)	O(n)
Two-pointer	O(n)	O(1) ✔️ Best
