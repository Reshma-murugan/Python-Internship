```python
def check_palindrome(num):
    ogNum = num
    rev = 0
    while num != 0:
        last_digit = num % 10
        rev = rev * 10 + last_digit
        num = num // 10
    if ogNum == rev:
        return True
    else:
        return False    

def palindrome_list(numbers):
    result = []
    for i in numbers:
        if check_palindrome(i):
            result.append(i)
    return result

numbers = [121, 133, 123, 88, 97]
print(palindrome_list(numbers))
```
