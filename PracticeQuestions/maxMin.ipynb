```python
def maxMin(array):
    max=array[0]
    min=array[0]
    for x in array:
        if x>max:
            max=x
        elif x<min:
            min=x
    return max,min

x=[1,2,3,4,5,6]
max,min=maxMin(x) 
print("max:",max,"min:",min)
# print(result) 
# print("max:",result[0],"; min:",result[1])
       
```

✅ Time Complexity → O(n)

Your function scans the entire list one time:

for x in array:

So the time grows linearly with the number of elements.

✅ Space Complexity → O(1)

You use only two extra variables:

maximum, minimum

No list, no dictionary, no new data structure.

So the space used is constant.

⭐ Final Answer
Time Complexity  = O(n)
Space Complexity = O(1)
