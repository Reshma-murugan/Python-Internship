```python
def count_vowels_consonants(s):
    vowel = 0
    consonant = 0
    
    for ch in s:
        # convert to lowercase on the fly
        c = ch.lower()
        if c in "aeiou":
            vowel += 1
        elif c.isalpha():
            consonant += 1
    
    return vowel, consonant

x = count_vowels_consonants("Apple")
print("vowelCount =", x[0], "consonantCount =", x[1])
```
