3.Prime Check

A prime number is a number that has exactly two factors:

1

The number itself

That means a prime number is divisible by only 1 and itself.
1 → NOT a prime number

(It has only one factor → 1)

2 → Only even prime number

(Every other even number is divisible by 2)


```python
def is_prime(n):
    if n<=1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    
    return True

x=is_prime(7)
if x:
    print("prime")
else:
    print("not prime")    

    
```

❌ If you use for(int i = 2; i <= n; i++)

You are checking every number up to n, which is unnecessary and slow.

Example:
Check if 1000 is prime
Your loop runs:

2, 3, 4, 5, ……, 1000   → 999 checks ❌

This is too long.

✅ Why for(int i = 2; i <= sqrt(n); i++) is enough?

Because a composite number always has at least one factor ≤ √n.

Example:

Number	Factors
100	10 × 10
36	6 × 6
49	7 × 7
63	7 × 9
1000	10 × 100

See the pattern:

👉 If a number has a factor above √n,
it must have a matching factor below √n.

So checking only up to √n is enough.

📌 Example: n = 100

√100 = 10

Factors:

4 × 25

5 × 20

10 × 10

The smallest factor is always ≤ 10.

So once you reach √n, you already know whether n has a factor.

⭐ Complexity Difference
Loop	Iterations	Time Complexity
i <= n	n steps ❌	O(n)
i <= √n	√n steps ✔	O(√n)

Example for n = 1,000,000:

Method	Iterations
i ≤ n	1,000,000 checks ❌
i ≤ √n	1000 checks ✔

Time Complexity = O(√n)
Because we check divisors only up to the square root of n.

Space Complexity = O(1)
Because we use only a constant number of variables.
