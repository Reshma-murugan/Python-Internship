1. Check for Anagram
Two words are anagrams if:

🔹 They have the same letters
🔹 Each letter appears the same number of times
🔹 Just arranged in a different order

```python
def is_anagram(s1,s2):
    if sorted(s1)==sorted(s2):
        return "anagram"
    else:
        return "not a anagram"

s1="listen"
s2="silent"
print(is_anagram(s1,s2))
```

```python
def is_anagram(s1, s2):
    if len(s1) != len(s2):
        return False
    
    count = [0] * 26

    for ch in s1:
        count[ord(ch) - ord('a')] += 1

    for ch in s2:
        count[ord(ch) - ord('a')] -= 1

    for x in count:
        if x != 0:
            return False


    return True


x = is_anagram("silent", "listen")

if x:
    print("anagram")
else:
    print("not an anagram")
```
