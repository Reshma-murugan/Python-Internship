```python
def remove_duplicates(lst):
    unique = []
    for x in lst:
        if x not in unique:
            unique.append(x)
    return unique


l = int(input("Enter the number of elements in the list: "))
lst = []

for i in range(l):
    n = int(input(f"Enter element {i+1}: "))
    lst.append(n)

result = remove_duplicates(lst)
print("List after removing duplicates:", result)
```
