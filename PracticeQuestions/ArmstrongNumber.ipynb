```python
def check_Armstrong(num):
    og = num
    temp = num
    count = 0

    while temp != 0:
        count += 1
        temp //= 10   # use temp here

    sum = 0
    temp = num       # reset temp to use again

    while temp != 0:
        digit = temp % 10
        sum = sum + (digit ** count)
        temp = temp // 10

    if og == sum:
        print("given number is armstrong number")
    else:
        print("given number is not an armstrong number")

num = int(input("enter a number: "))
check_Armstrong(num)
```
