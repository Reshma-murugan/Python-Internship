```python
def Find_Largest(a,b,c):
    if a>b and a>c:
        print(a,"is largest")
    elif b>c:
        print(b,"is largest") 
    else:
        print(c,"is largest")    

a=int(input("enter first number"))
b=int(input("enter second number"))
c=int(input("enter third number"))
Find_Largest(a,b,c)
```
