```python
class calculator():

    def add(add):
        print(a+b)
    def sub(sub):
        print(a-b)
    def mul(mul):
        print(a*b)
    def div(div):
        print(a//b)
obj1=calculator()
obj1.add(10,5)
obj1.sub(10,7)
obj1.mul(10,2)
obj1.div(10,3)
```
