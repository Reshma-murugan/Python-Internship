A1. Even or Odd

```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

A2. Print Grade (A/B/C/D/F)


```python
marks = int(input("Enter marks: "))

if marks >= 90:
    print("A")
elif marks >= 75:
    print("B")
elif marks >= 60:
    print("C")
elif marks >= 45:
    print("D")
else:
    print("F")
```

A3. Positive, Negative, or Zero

```python
num = float(input("Enter a number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

A4. Leap Year

```python
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

A5. Smallest of Three Numbers

```python
a = int(input("Enter number 1: "))
b = int(input("Enter number 2: "))
c = int(input("Enter number 3: "))

if a < b and a < c:
    print("Smallest:", a)
elif b < c:
    print("Smallest:", b)
else:
    print("Smallest:", c)
```

A6. Password Validator (Password → Python123)

```python
password = input("Enter password: ")

if password == "Python123":
    print("Access Granted")
else:
    print("Access Denied")
```

A7. Simple Calculator (+, -, *, /)


```python
a = float(input("Enter number 1: "))
b = float(input("Enter number 2: "))
op = input("Enter operator (+, -, *, /): ")

if op == "+":
    print(a + b)
elif op == "-":
    print(a - b)
elif op == "*":
    print(a * b)
elif op == "/":
    print(a / b)
else:
    print("Invalid operator")
```

A8. Voting Eligibility (age ≥ 18)



```python
age = int(input("Enter age: "))

if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible to vote")
```

A9. Character Type (Vowel / Consonant / Other)

```python
ch = input("Enter a character: ").lower()

if ch in "aeiou":
    print("Vowel")
elif ch.isalpha():
    print("Consonant")
else:
    print("Other")
```

A10. Weekend or Weekday


```python
day = input("Enter a day: ").lower()

if day == "saturday" or day == "sunday":
    print("Weekend")
else:
    print("Weekday")
```

SECTION C – String Operations

Given:

s = "HelloWorld"

C1. Insert "Python" at 5th index

```python
s = "HelloWorld"
s = s[:5] + "Python" + s[5:]
print(s)
```

```python
s = s.replace("World", "")
print(s)
```

C3. Append "2024"

```python
s = s + "2024"
print(s)
```

```python
s = s[:2] + s[3:]
print(s)
```
