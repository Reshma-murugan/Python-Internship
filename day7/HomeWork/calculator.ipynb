```python
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))

c = int(input("Select the operation number: \n1 for Addition \n2 for Subtraction \n3 for Multiplication \n4 for Division\nEnter your choice: "))

if c == 1:
    print("a + b =", a + b)

elif c == 2:
    print("a - b =", a - b)

elif c == 3:
    print("a * b =", a * b)

elif c == 4:
    if b == 0:
        print("Error: Cannot divide by zero.")
    else:
        print("a / b =", a / b)

else:
    print("Invalid choice")
```
