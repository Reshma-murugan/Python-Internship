```python
L=[59,1,89,30,2,5]
min=L[0]
max=L[0]
for x in L:
    if x>max:
        max=x
    elif x<min:
        min=x
print("max: ",max)
print("min: ",min)        
```

```python
L=[59,1,89,30,2,5]
print("max=",max(L))
print("min=",min(L))
```

secondLargest

```python
L=[59,1,89,30,2,5]
firstLarge=float('-inf')
secondLarge=float('-inf')
for x in L:
    if x>firstLarge:
        secondLarge=firstLarge
        firstLarge=x
    elif x<firstLarge and x>secondLarge:
        secondLarge=x
print("secondLargst number in an array = ",secondLarge)
```

```python
L = [59, 1, 89, 30, 2, 5]
sorted_list=sorted(L)            # ascending order
secondLargest=sorted_list[-2]    # second from the end

print("secondLargest =", secondLargest)
```

SecondSmallest

```python
L=[59,1,89,30,2,5]
firstSmallest=float('inf')
secondSmallest=float('inf')
for x in L:
    if x<firstSmallest:
        secondSmallest=firstSmallest
        firstSmallest=x
    elif x>firstSmallest and x<secondSmallest:
        secondSmallest=x
print("secondSmallest=",secondSmallest)
```

```python
L = [59, 1, 89, 30, 2, 5]
sorted_list=sorted(L)            # ascending order
secondSmallest=sorted_list[1]    # second from the start

print("secondSmallest =", secondSmallest)
```
