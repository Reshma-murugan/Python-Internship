```python
words = ["Python", "is", "fun"] #Join words using a space
sentence = " ".join(words)
print(sentence)
```

```python
txt = "Thank you for the music\nWelcome to the jungle" #splitlines() splits a string wherever it finds a newline (\n, \r, or both) and returns a list of lines.
x = txt.splitlines()
print(x)
```

```python
txt = "50" #Make the total length 10 characters,and fill the left side with zeros.
x = txt.zfill(10)
print(x)
```
