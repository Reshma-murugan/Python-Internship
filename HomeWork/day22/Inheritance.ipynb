Inheritance is an object-oriented programming concept where one class (child class) can use the properties and methods of another class (parent class).

1. Single Inheritance

Single inheritance means one child class inherits from one parent class.

```python
class University:
    university = "Anna University"

class College(University):
    college = "PET Engineering College"

obj = College()

u = obj.university
c = obj.college

print(f"I completed my degree in {c} under {u}")
```

2. Multiple Inheritance

Multiple inheritance means one child class inherits from more than one parent class.

```python
class University:
    university = "Anna University"

class Course:
    course = "B.E"

class College(University, Course):
    college = "PET Engineering College"

obj = College()

u = obj.university
c = obj.college
co = obj.course

print(f"I completed {co} in {c} under {u}")
```

3. Multilevel inheritance 

means a class inherits from another child class.
where a class (child) inherits from a derived class (parent), which in turn inherits from a base class (grandparent). This creates a chain (A->B->C) 

```python
class University:
    university = "Anna University"

class College(University):
    college = "PET Engineering College"

class Course(College):
    course = "B.E"

class Department(Course):
     department="Computer Science"

class Student(Department):
    student="Reshma"     

obj1 = Course()
obj2 = Student()


print(f"I completed {obj1.course} in {obj1.college} under {obj1.university}")
print(f"I am {obj2.student}, I completed {obj1.course}-{obj2.department} in {obj1.college} under {obj1.university}")
```

4. Hierarchical Inheritance

Hierarchical inheritance means multiple child classes inherit from one parent class.

```python
class University:
    university = "Anna University"

class Department1(University):
    department = "B.E"

class Department2(University):
    department = "M.E"

obj1 = Department1()
obj2 = Department2()

print(f"{obj1.department} students studies under {obj1.university}")
print(f"{obj2.department} students studies under {obj2.university}")
```

```python
class University:
    university = "Anna University"

class College(University):
    college = "PET Engineering College"

class Course(University):
    course = "B.E"

class Department(College, Course):
    department = "Computer Science"

obj = Department()

print(f"I completed {obj.course}-{obj.department} in {obj.college} under {obj.university}")
```
