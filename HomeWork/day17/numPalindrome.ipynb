```python
def is_palindrome(n): 
    start=0    
    end=len(n)-1 
    while start<=end:
        if n[start]!=n[end]:
            return "not a palindrome"
        start+=1
        end-=1
    return "palindrome"   

n=input("enter the number") 
print(is_palindrome(n))
```

✅ Time Complexity → O(n)

In each iteration, you compare one character from the start and one from the end.

You move inward until you cover half the string.

Even though you stop at half, we still say O(n) because Big-O ignores constants.

So time complexity = O(n).

✅ Space Complexity → O(1)

You use only two extra variables: start and end.

No new array or string copy is created.

Extra memory does not grow with input size.

So space complexity = O(1).

```python
def is_palindrome(n):
    rev=0
    og_num=n
    while n!=0:
        digit=n%10
        rev=rev*10+digit
        n=n//10
    if(rev==og_num):
        return "palindrome"    
    else:
        return "not a palindrome"   
n=int(input("enter a number"))   
print(is_palindrome(n))      
```
