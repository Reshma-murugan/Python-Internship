```python
s = "hello"
print(int(s)) 
'''ValueError: invalid literal for int() with base 10: 'hello'
Because "hello" is not a number, so it cannot be converted to int.'''


```

```python
s = "hello"
print(float(s)) #ValueError: could not convert string to float: 'hello'
```

```python
s = "hello"
print(bool(s))  #True - Because "hello" is a non-empty string, so Python treats it as True.
print(complex(s)) 
'''ValueError: complex() arg is a malformed string
Python cannot convert "hello" to a complex number.'''
```

```python
c = 10+5j
print(int(c))
'''Real part → 10

Imaginary part → 5

Python doesn’t know which one you want to convert to integer, so direct conversion is not allowed.'''
```

```python
c = 10 + 5j

print(int(c.real))   # 10
print(int(c.imag))   # 5
```

```python
c = 10 + 5j
print(str(c)) #Python can always convert any value to a string.
print(bool(c)) #any non-zero number including complex → True Only 0 + 0j → False
print(float(c))
```

```python
c = 10 + 5j

print(float(c.real))   # 10.0
print(float(c.imag))   # 5.0
```
