```python
employees = {
    1: ["Arun", 20000],
    2: ["Bala", 25000],
    3: ["Reshma", 30000],
    4: ["Kowsalya", 28000],
    5: ["Esther", 27000],
    6: ["Indhu", 26000],
    7: ["Mauni", 35000],
    8: ["Hari", 32000],
    9: ["Uma", 29000],
    10: ["Jana", 31000]
}

print("Enter your choice: \n1. Add \n2. Edit \n3. Delete")

choice = int(input())


# ---------------- ADD ----------------
if choice == 1:
    id = int(input("Enter new ID: "))
    name = input("Enter name: ")
    salary = float(input("Enter salary: "))
    employees[id] = [name, salary]
    print("Employee added!")


# ---------------- EDIT ----------------
elif choice == 2:
    old = input("Enter old name: ")

    found = False

    if employees[1][0] == old:
        employees[1][0] = input("Enter new name: ")
        found = True
    elif employees[2][0] == old:
        employees[2][0] = input("Enter new name: ")
        found = True
    elif employees[3][0] == old:
        employees[3][0] = input("Enter new name: ")
        found = True
    elif employees[4][0] == old:
        employees[4][0] = input("Enter new name: ")
        found = True
    elif employees[5][0] == old:
        employees[5][0] = input("Enter new name: ")
        found = True
    elif employees[6][0] == old:
        employees[6][0] = input("Enter new name: ")
        found = True
    elif employees[7][0] == old:
        employees[7][0] = input("Enter new name: ")
        found = True
    elif employees[8][0] == old:
        employees[8][0] = input("Enter new name: ")
        found = True
    elif employees[9][0] == old:
        employees[9][0] = input("Enter new name: ")
        found = True
    elif employees[10][0] == old:
        employees[10][0] = input("Enter new name: ")
        found = True

    if found:
        print("Name updated!")
    else:
        print("Record not found")


# ---------------- DELETE ----------------
elif choice == 3:
    name = input("Enter name to delete: ")

    deleted = False

    if employees[1][0] == name:
        del employees[1]; deleted = True
    elif employees[2][0] == name:
        del employees[2]; deleted = True
    elif employees[3][0] == name:
        del employees[3]; deleted = True
    elif employees[4][0] == name:
        del employees[4]; deleted = True
    elif employees[5][0] == name:
        del employees[5]; deleted = True
    elif employees[6][0] == name:
        del employees[6]; deleted = True
    elif employees[7][0] == name:
        del employees[7]; deleted = True
    elif employees[8][0] == name:
        del employees[8]; deleted = True
    elif employees[9][0] == name:
        del employees[9]; deleted = True
    elif employees[10][0] == name:
        del employees[10]; deleted = True

    if deleted:
        print("Employee deleted!")
    else:
        print("Record not found")

else:
    print("Invalid choice")
```
