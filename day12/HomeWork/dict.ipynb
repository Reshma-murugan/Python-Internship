```python
employees = {
    1: ["Arun", 200000],
    2: ["Bala", 200000],
    3: ["Reshma", 50000],
    4: ["Kowsalya", 28000],
    5: ["Esther", 27000],
    6: ["Indhu", 26000],
    7: ["Mauni", 35000],
    8: ["Hari", 32000],
    9: ["Uma", 29000],
    10: ["Jana", 31000]
}

print("Enter your choice:\n1.Add\n2.Edit\n3.Delete\n4.View All")
choice = int(input("enter your choice"))


# -------- ADD --------
if choice == 1:
    id = int(input("Enter ID: "))
    name = input("Enter name: ")
    salary = float(input("Enter salary: "))
    employees[id] = [name, salary]
    print("Employee added!")


# -------- EDIT --------
elif choice == 2:
    old = input("Enter old name: ")
    found = False

    for key in employees:
        if employees[key][0] == old:
            employees[key][0] = input("Enter new name: ")
            found = True
            break

    if found:
        print("Name updated!")
    else:
        print("Record not found")


# -------- DELETE --------
elif choice == 3:
    name = input("Enter name to delete: ")
    deleted = False

    for key in employees:
        if employees[key][0] == name:
            del employees[key]
            deleted = True
            break

    if deleted:
        print("Employee deleted!")
    else:
        print("Record not found")

# -------- VIEW ALL --------

elif choice == 4:
    for key in employees:
        print("ID:", key, "Name:", employees[key][0], "Salary:", employees[key][1])        

else:
    print("Invalid choice")
```
