```python
skill = input("Enter skill: ").lower()
location = input("Enter your location: ").lower()
experience = input("Enter your experience / fresher: ").lower()

# Normalize experience
if "1" in experience:
    experience = "1 year"
    
    

if "0" in experience:
    experience = "fresher"

if skill == "python" and location == "chennai" and experience == "1 year":
    print("You are eligible for Infosys")

elif skill == "python" and location == "coimbatore" and experience == "fresher":
    print("You are eligible for Wipro")

elif skill == "java" and location == "chennai" and experience == "fresher":
    print("You are eligible for TCS")

elif skill == "java" and location == "bangalore" and experience == "1 year":
    print("You are eligible for Cognizant")

else:
    print("You are not eligible")
```
